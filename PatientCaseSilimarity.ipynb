```python
from sklearn.preprocessing import StandardScaler

# create a scaler object
scaler = StandardScaler()

# standardize the numerical attributes
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)


# create a range of cluster numbers to try
n_clusters = range(1, 11)

# create a list to store the inertia values
inertia_values = []

# fit KMeans models for each value of n_clusters and record the inertia
for k in n_clusters:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# plot the elbow curve to select the optimal number of clusters
plt.plot(n_clusters, inertia_values, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()


```

```python
# select the number of clusters based on the elbow curve
num_clusters = 4

# create a KMeans object with 4 clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=0)

# fit the KMeans model on the preprocessed data
kmeans.fit(X_scaled)

# obtain cluster labels for each data point
cluster_labels = kmeans.labels_
```

```python
df['cluster'] = cluster_labels

# summarize each cluster's characteristics
cluster_summary = df.groupby('cluster')[features].mean().reset_index()

# add the number of patients in each cluster to the summary
cluster_counts = df['cluster'].value_counts().reset_index().rename(columns={'index': 'cluster', 'cluster': 'count'})
cluster_summary = cluster_summary.merge(cluster_counts, on='cluster')

# print the cluster summary
print(cluster_summary)
```
